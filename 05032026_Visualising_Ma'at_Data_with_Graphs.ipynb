```
# 1. Install the required network physics library
!pip install pyvis -q

import json
from pyvis.network import Network
import IPython

# 2. Embed the data
data_json = """
{
  "name": "mꜣꜥt", "gloss": "Ma'at (Cosmic Order)", "category": "Spirituality", "epochs": ["Egypt", "Nubia", "Aksum", "Mali"],
  "description": "The fundamental principle of existence: truth, balance, order, law, and justice.",
  "children": [
    {
      "name": "nsw", "gloss": "The King (Authority)", "category": "Law", "epochs": ["Egypt", "Nubia", "Aksum", "Mali"],
      "description": "The central pivot who maintains Ma'at on earth.",
      "children": [
        {
          "name": "išd", "gloss": "Ished Tree", "category": "Law", "epochs": ["Egypt", "Nubia"],
          "description": "The persea tree where the king's name and years were recorded by Thoth and Seshat.",
          "children": [
             {"name": "rnpt", "gloss": "Palm Rib (Year/Time)", "category": "Law", "epochs": ["Egypt"], "description": "Symbol of long reign and the measurement of time."},
             {"name": "ngś ngśt", "gloss": "King of Kings", "category": "Law", "epochs": ["Aksum"], "description": "Ezana's title asserting suzerainty over diverse tribes (Noba, Kasu) and unifying the Horn."},
             {"name": "mansa", "gloss": "Mansa (Conqueror/King)", "category": "Law", "epochs": ["Mali"], "description": "Sundiata's title as the unifier of the Mandé peoples."}
          ]
        },
        {
          "name": "hp", "gloss": "Law/Custom", "category": "Law", "epochs": ["Egypt", "Mali"],
          "description": "The codification of social rules.",
          "children": [
            {"name": "Kouroukan Fouga", "gloss": "Manden Charter", "category": "Social", "epochs": ["Mali"], "description": "Sundiata's constitution declaring social peace and the sanctity of life."}
          ]
        }
      ]
    },
    {
      "name": "nḥt", "gloss": "Sycamore (Regeneration)", "category": "Spirituality", "epochs": ["Egypt", "Nubia"],
      "description": "The tree of the goddess (Hathor/Nut) providing water and bread to the dead.",
      "children": [
        {
          "name": "nw", "gloss": "Cool Water", "category": "Spirituality", "epochs": ["Egypt", "Nubia"],
          "description": "Life-force offered to the Ba soul.",
          "children": [
            {"name": "Egziabher", "gloss": "Lord of Heaven", "category": "Spirituality", "epochs": ["Aksum"], "description": "Ezana's monotheistic shift; water/font becomes baptismal and purifying in a Christian context."}
          ]
        },
        {
          "name": "prt", "gloss": "Fruit/Sustenance", "category": "Environment", "epochs": ["Egypt", "Mali"],
          "description": "Agricultural abundance as proof of divine favor.",
          "children": [
            {"name": "Niger/Sankarani", "gloss": "River Valleys", "category": "Environment", "epochs": ["Mali"], "description": "The agricultural heartland secured by Sundiata for rice and grain production."},
            {"name": "Terraces", "gloss": "Highland Agriculture", "category": "Environment", "epochs": ["Aksum"], "description": "Ezana's era saw intensive adaptation of the environment for wheat/teff."}
          ]
        }
      ]
    },
    {
      "name": "św", "gloss": "Trade/Exchange", "category": "Commerce", "epochs": ["Egypt", "Aksum", "Mali"],
      "description": "Economic Ma'at: fair exchange and safe passage.",
      "children": [
        {"name": "Adulis", "gloss": "Port of Adulis", "category": "Commerce", "epochs": ["Aksum"], "description": "The gateway connecting Rome, India, and Africa (Red Sea trade)."},
        {"name": "Wangara", "gloss": "Gold Fields", "category": "Commerce", "epochs": ["Mali"], "description": "The source of wealth that Sundiata secured, fueling trans-Saharan trade."}
      ]
    },
    {
      "name": "isfet", "gloss": "Isfet (Chaos)", "category": "Spirituality", "epochs": ["Egypt", "Mali", "Aksum"],
      "description": "The opposite of Ma'at: violence, injustice, and lies.",
      "children": [
        {"name": "Sumanguru", "gloss": "Sorcerer King", "category": "Social", "epochs": ["Mali"], "description": "The antagonist to Sundiata; represented chaos, dark magic, and tyranny."},
        {"name": "Noba/Kasu", "gloss": "Rebellious Tribes", "category": "Social", "epochs": ["Aksum"], "description": "Forces of disorder that Ezana had to subdue to restore peace."}
      ]
    }
  ]
}
"""

personas_json = """
{
  "ezana": {"color": "#00bcd4", "name": "Ezana of Aksum", "nodes": ["ngś ngśt", "Egziabher", "Adulis", "Terraces", "Noba/Kasu", "nsw", "mꜣꜥt"]},
  "sundiata": {"color": "#ffc107", "name": "Sundiata Keita", "nodes": ["mansa", "Kouroukan Fouga", "Wangara", "Niger/Sankarani", "Sumanguru", "hp", "nsw", "mꜣꜥt"]}
}
"""

data = json.loads(data_json)
personas = json.loads(personas_json)

# 3. Initialize the Pyvis Network with a dark theme
net = Network(height="600px", width="100%", bgcolor="#121212", font_color="#e0e0e0", notebook=True, cdn_resources='remote')
net.force_atlas_2based(gravity=-75, central_gravity=0.01, spring_length=150) # The physics engine

sundiata_nodes = set(personas["sundiata"]["nodes"])
ezana_nodes = set(personas["ezana"]["nodes"])

# 4. Recursive function to parse JSON and map it to graph nodes/edges
def add_nodes_edges(node, parent_name=None):
    node_id = node['name']
    label = f"{node['name']}\n({node['gloss']})"

    # Format the hover tooltip info
    title = f"Epochs: {', '.join(node.get('epochs', []))}\nCategory: {node.get('category', 'N/A')}\n\n{node.get('description', '')}"

    # Determine color based on personas (Shared nodes get a purple mix)
    if node_id in sundiata_nodes and node_id in ezana_nodes:
        node_color = "#9c27b0" # Shared (e.g., mꜣꜥt, nsw)
    elif node_id in sundiata_nodes:
        node_color = personas["sundiata"]["color"]
    elif node_id in ezana_nodes:
        node_color = personas["ezana"]["color"]
    else:
        node_color = "#333333" # Default unassigned

    # Make the root node (Ma'at) slightly larger
    node_size = 25 if not parent_name else 15

    # Add the node to the graph
    net.add_node(node_id, label=label, title=title, color=node_color, size=node_size, shape="dot")

    # If there's a parent, draw the connecting spring (edge)
    if parent_name:
        net.add_edge(parent_name, node_id, color="#444444")

    # Recurse for children
    if 'children' in node:
        for child in node['children']:
            add_nodes_edges(child, node_id)

add_nodes_edges(data)

# Optional: Shows sliders below the graph so you can adjust the gravity/physics in real-time
net.show_buttons(filter_=['physics'])

# 5. Generate and render the interactive graph inline in Colab
net.show("maat_network.html")
IPython.display.HTML("maat_network.html")
```
